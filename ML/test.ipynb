```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

data = {
    'Date': pd.date_range(start='2023-03-17', periods=20*17, freq='H'),  # 날짜 정보 (시작 날짜와 주기, n은 주차 수)
    'Hour': list(range(6, 23)) * 20,  # 시간 정보 (6시부터 22시까지)
    'Passengers': [0] * 17 * 20,  # 해당 시간의 이용객 수
}

df = pd.DataFrame(data)

df.set_index('Date', inplace=True)

train_data = df['Passengers'][:-17]
test_data = df['Passengers'][-17:]

model = ARIMA(train_data, order=(3,1,1))
model_fit = model.fit()

predicted_passengers = model_fit.forecast(steps=17)

next_week_dates = pd.date_range(start=df.index[-1] + pd.DateOffset(hours=1), periods=17, freq='H')
result = pd.DataFrame({'Date': next_week_dates, 'Predicted_Passengers': predicted_passengers})
print(result)
```
